```python
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill", palette=["m", "g","b"],
            data=tips)
sns.despine(offset=10, trim=True)
```

```python
# Draw a nested boxplot to show bills by day and time
g1 = sns.boxplot(x="Database", y="Value", palette=["m", "g","b"],
            data=gifsOne)
g1.set_title("Internal Latency (one gif)")
g1.set_ylabel("Latency (ms)",fontsize=10)
g1fig = g1.get_figure()
g1fig.savefig("LatencyDatabases_oneGif.png", bbox_inches='tight')
```
